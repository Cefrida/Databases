## Start Part 1

```python
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt

f, axes = plt.subplots(2, 2, figsize=[14, 10], gridspec_kw={'wspace': 0.3,'hspace':0.4})
axes = axes.flatten()

legend_title = "University Rank"
y_var = 'total_score'
y_labels = 'Total Score'
x_vars = ['num_students', 'student_staff_ratio', 'international_students', 'female']
x_labels = ['Count of Total Number of Students', 'Student to Staff Ratio', 'Percent of International Students', 'Percent of Female Students']

# Loop to create the plots
for i in range(len(axes)):
    x_var = x_vars[i]
    
    sns.scatterplot(data=top_200, x=x_var, y=y_var, s=90, alpha=0.8, ec='k',
                    ax=axes[i], label='Top 200', marker='o', color = 'lightgray')
    sns.scatterplot(data=top_10, x=x_var, y=y_var, s=90, alpha=0.8, ec='k',
                    ax=axes[i], label='Top 10', marker='o', color = 'firebrick')
   

    # Set x-axis and y-axis labels
    axes[i].set_xlabel(x_labels[i], fontsize=14, labelpad=15)
    axes[i].set_ylabel(y_labels, fontsize=14, labelpad=15)
    axes[i].tick_params(labelsize=12)

    if i != 1:
        axes[i].get_legend().remove()
    else:
        # Create a legend for the second plot
        axes[i].legend(fontsize=13, title=legend_title, markerfirst=False, ncol=3,
               bbox_to_anchor=[0.5, 0.95], markerscale=1.5, columnspacing=1,
               handletextpad=0, facecolor='whitesmoke', loc='center',bbox_transform = f.transFigure,
               title_fontsize=13)


    # Calculate the median of the x values
    median_x = top_200[x_var].median()

    # Add a vertical dashed line at the median
    axes[i].axvline(median_x, linestyle='--', color='red', alpha=0.5)
    
 

plt.savefig('scatter.png', bbox_inches='tight', facecolor='white')
plt.show()
```

```python
#Correlation Coefficients 
correlation_coefficients = college[x_vars].corrwith(college[y_var])

# Print the correlation coefficients
print(correlation_coefficients)
```

## Start Part 2

```python
last_200= college.iloc[-200:]
last_200
```

```python
first_last = pd.concat([top_200, last_200]).reset_index()
```

```python
first_last['Ranks'] = ['top']*200 + ['last']*200
```

```python
first_last
```

```python
from pandas.plotting import parallel_coordinates as pcp
from matplotlib.patches import Patch
```

```python
#first filter columns 
variables = ['teaching','income', 'citations',  'research','international', 'female', 'Ranks']

# Create a new DataFrame with selected columns
new_df = first_last[variables].copy()
new_df
column_name_mapping = {
    'income': 'Income',
    'teaching': 'Teaching',
    'citations': 'Citations',
    'research': 'Research',
    'international': 'International',
    'female': 'Female'
}

# Rename columns
new_df.rename(columns=column_name_mapping, inplace=True)
```

```python
plt.figure(figsize = [12,8])
colors = ['dodgerblue','orangered']
pcp(new_df, 'Ranks', color = colors, alpha = 0.5, axvlines_kwds={'alpha':0.5,'color':'grey'})
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.grid(axis = 'y', alpha = 0.4, linestyle = '--')
#How to make a fake legend

handles = [Patch(facecolor=c, alpha = 0.6) for c in colors]
plt.legend(handles, ['Top 200','Last 200'], title='College Rank',
             fontsize = 11, title_fontsize = 11, frameon = True, loc='upper right')
plt.ylabel('Percentage', fontsize = 13, labelpad = 20, fontweight="bold")
plt.savefig('parallel.png',bbox_inches ='tight',facecolor='whitesmoke')
plt.show()
```

```python
college['column1_log'] = college['num_students'].apply(np.log10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

f, axes = plt.subplots(2, 2, figsize=[14, 10], gridspec_kw={'wspace': 0.2, 'hspace': 0.3})
axes = axes.flatten()

y_var = 'total_score'
y_label = 'Total Score'
x_vars = ['column1_log', 'student_staff_ratio', 'international_students', 'female']
x_labels = ['Count of Total Number of Students (log base 10)', 'Student to Staff Ratio', 'Percent of International Students', 'Percent of Female Students']



# Loop to create the hexbin plots
for i in range(len(axes)):
    x_var = x_vars[i]

    hb = axes[i].hexbin(college[x_var], college[y_var], gridsize=15, cmap='magma_r', mincnt = 1)
    plt.colorbar(hb, ax=axes[i])
    
    axes[i].set_xlabel(x_labels[i], fontsize=13, labelpad=15)
    axes[i].set_ylabel(y_label, fontsize=13, labelpad=15)
    axes[i].tick_params(labelsize=11)

plt.savefig('hex.png', bbox_inches='tight', facecolor='white')
plt.show()
```
